## Q1. Student Class

Create a class `Student` with name, age and course. Create an object and display all details.

```
Sample Input:
Name = Rahul
Age = 20
Course = Python

Sample Output:
Name : Rahul
Age : 20
Course : Python
```

````python
# Write your code here
class Student:
    def __init__(self,name,age,course):
        self.name=name
        self.age=age
        self.course=course
    def showdetails(self):
        print('Name :',self.name)  
        print("Age :",self.age)
        print("Course :",
        
        self.course)  
obj=Student("Rahul",20,"B.tech")
obj.showdetails()        
````


## Q9. Rectangle Class

Create a class `Rectangle` with length and breadth. Create method `area()`.

```
Sample Input:
Length = 10
Breadth = 5

Sample Output:
Area = 50
```

````python
# Write your code here
class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth

    def area(self):
        print("Area = ",self.length*self.breadth)

obj=Rectangle(10,5)
obj.area()            
````

## Q10. Circle Class

Create a class `Circle` with radius. Create method `area()`. Use 3.14*r*r.

```
Sample Input:
Radius = 7

Sample Output:
Area = 153.86
```

````python
# Write your code here
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        print("Area = ",self.radius*self.radius*3.14)

obj=Circle(7) 
obj.area()           
````
