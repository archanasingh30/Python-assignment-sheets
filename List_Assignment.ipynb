### Q4. Count Even Numbers
```python
nums = [12, 5, 8, 7, 10, 13, 16]
```
Count how many even numbers are present.

**Expected Output**
```python
Total Even Numbers = 4
```


````python
# Write your code here
nums=[12,5,8,7,10,13,16]
even=0
for i in range(len(nums)):
    if nums[i]%2==0:
        even+=1
print(even)
````

### Q8. Create a New List of Squares

### Q9. Count Positive and Negative Numbers
```python
nums = [10, -5, 20, -8, 15, -2]
```
Count positive and negative numbers separately.

**Expected Output**
```python
Positive = 3
Negative = 3
```


````python
# Write your code here
nums=[10,-5,20,-8,15,-2]
positive_count=0
negative_count=0
for i in range(len(nums)):
    if nums[i]>=0:
        positive_count+=1
    else:
        negative_count+=1 
print(f"Positive = ",positive_count) 
print(f"Negative = ",negative_count)         
````

### Q10. Second Largest Number
```python
nums = [12, 45, 23, 89, 67]
```
Find the second largest number.

**Expected Output**
67
```


````python
# Write your code here
nums=[12,45,23,89,67]
largest=nums[0]
second_largest=largest
for i in range (len(nums)):
    if nums[i]>largest:
        second_largest=largest
        largest=nums[i]
    elif nums[i]>second_largest:
        second_largest=nums[i]    
print(second_largest)        
````
